I am going to make a table with uninsured rates between 2012 and 2017 so I can compare them more easily.

```python
df_joined_2012_2017 = df_2017_states.sort_values(by='Uninsured', ascending=True).join(df_2012_states.sort_values(by='Uninsured', ascending=False),lsuffix="_2017",rsuffix='_2012').head(52)
```

```python
df_joined_2012_2017 = df_joined_2012_2017.drop(['Footnotes_2017', 'Footnotes_2012'],axis=1)
```

```python
df_joined_2012_2017['diff']= df_joined_2012_2017['Uninsured_2017'] - df_joined_2012_2017['Uninsured_2012']
```

```python
df_joined_2012_2017
```

```python
def label_diff(difference):
    if difference > 0:
        return 'positive'
    if difference == 0:
        return 'no change'
    if difference < 0:
        return 'negative'
```

```python
df_joined_2012_2017['labeled'] = df_joined_2012_2017['diff'].apply(lambda x: label_diff(x))
```

```python
df_joined_2012_2017['labeled'].value_counts()
```

```python
#uninsurance rates decreased across the board, except in Puerto Rico.
```

```python
df_medicaid_expansion['Description'].str.extract(r'.*\d+/\d+/(\d\d\d\d)')
```

```python
df_medicaid_expansion['Implementation'] = df_medicaid_expansion['Description'].str.extract(r'.*\d+/\d+/(\d\d\d\d)')
```

```python
df_medicaid_expansion
```

```python
df_medicaid_expansion['Expansion Status'].value_counts()
```

```python
#Fix Idaho and Virginia - it picked up enrollment rate and not implementation date
#Also fix Nebraska - not nan, but unknown (adopted but not implemented)
df_medicaid_expansion.at[12,'Implementation'] = '2020'
df_medicaid_expansion.at[46,'Implementation'] = '2019'
df_medicaid_expansion.at[27,'Implementation'] = 'unknown'
```

```python
df_medicaid_expansion['Implementation'].value_counts(dropna=False)
```

States w/o expansion

```python
list_states_no_expansion = df_medicaid_expansion[df_medicaid_expansion['Implementation'].isnull()]['State'].to_list()
```

```python
list_states_no_expansion
```

```python
list_states_no_implementation_2017 = df_medicaid_expansion[(df_medicaid_expansion['Implementation'].isnull())|(df_medicaid_expansion['Implementation']=='2018')|(df_medicaid_expansion['Implementation']=='2019')|(df_medicaid_expansion['Implementation']=='unknown')]['State'].to_list()



```

```python
len(list_states_no_implementation_2017)
```

# Comparing Medicaid expansion with states that are behind

```python
def expanded_classify(name):
    if name in list_states_no_expansion:
        return 'no'
    else:
        return 'investigate further'
```

```python
df_2017_states['expanded_boolean'] = df_2017_states.sort_values(by='Uninsured', ascending=False)['Location'].apply(lambda x: expanded_classify(x))
```

```python
df_2017_states['expanded_boolean'].value_counts()
```

```python
df_2017_states.sort_values(by='Uninsured', ascending=False).head(26)
```

```python
#13 out of 14 rank in the bottom 
```

```python
def expanded_2017(name):
    if name in list_states_no_implementation_2017:
        return 'no'
    else:
        return 'investigate further'
```

```python
df_2017_states['expanded_boolean_2017'] = df_2017_states.sort_values(by='Uninsured', ascending=False)['Location'].apply(lambda x: expanded_2017(x))
```
